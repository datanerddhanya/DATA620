```python
import networkx as nx
import matplotlib.pyplot as plt

# Creating an empty graph
kite_graph = nx.Graph()

# Adding nodes (0 through 9)
kite_graph.add_nodes_from(range(10))

# Defining the edges for the Krackhardt kite graph
edges = [
    (0, 1), (0, 2), (0, 3), (0, 5),
    (1, 3), (1, 4), (1, 6),
    (2, 3), (2, 5),
    (3, 4), (3, 5), (3, 6),
    (4, 6),
    (5, 6), (5, 7),
    (6, 7),
    (7, 8),
    (8, 9)
]

# Adding edges to the graph
kite_graph.add_edges_from(edges)

# Defining a dictionary for node labels
node_labels = {
    0: 'Jane', 1: 'Carl', 2: 'Heather', 3: 'Fernando', 4: 'Olivia',
    5: 'Alice', 6: 'Bob', 7: 'Diana', 8: 'Gaston', 9: 'Ed'
}

# layout algorithm to position the nodes
pos = nx.spring_layout(kite_graph)

# graph with the specified layout and labels
nx.draw(kite_graph, pos, labels=node_labels, with_labels=True, node_color='skyblue', node_size=700, edge_color='gray')
plt.title("Krackhardt Kite Graph with Spring Layout")
plt.show()
```
